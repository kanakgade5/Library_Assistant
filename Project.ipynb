```python
# Step 2: Import libraries
from transformers import pipeline
import gradio as gr
```

```python
# Step 3: Load AI model
chatbot = pipeline("text-generation", model="gpt2-medium")
```

```python

# Step 4: Knowledge base (FAQ + library info)
faq = {
    "library hours": "Library is open from 9 AM to 8 PM.",
    "exam schedule": "Mid-term exams start from 15th June.",
    "assignment deadline": "Submit assignments by 10th June.",
    "contact admin": "You can email admin@college.edu for queries."
}
```

```python
# Step 5: Book database
books = [
    {"title": "Artificial Intelligence Basics", "author": "Tom Taulli"},
    {"title": "Python for Data Science", "author": "Jake VanderPlas"},
    {"title": "Machine Learning Guide", "author": "Andrew Ng"}
]

def search_books(query):
    results = [b for b in books if query.lower() in b["title"].lower()]
    if results:
        return "Books found:\n" + "\n".join([f"{b['title']} by {b['author']}" for b in results])
    else:
        return "No books found for your query."
```

```python
# Step 6: Reminder system
reminders = []

def add_reminder(task, date):
    reminders.append({"task": task, "date": date})
    return f"Reminder added: {task} on {date}"
```

```python
# Step 7: Chat function
conversation_history = ""

def chat_with_bot(user_input):
    global conversation_history

    # Check FAQ
    for key in faq:
        if key in user_input.lower():
            return faq[key]

    # Check book search
    if "find book" in user_input.lower():
        query = user_input.lower().replace("find book", "").strip()
        return search_books(query)

    # Check reminder
    if "remind me" in user_input.lower():
        parts = user_input.split("on")
        if len(parts) == 2:
            task = parts[0].replace("remind me", "").strip()
            date = parts[1].strip()
            return add_reminder(task, date)
        else:
            return "Please say like: Remind me assignment on 10th June"

    # Otherwise use AI model
    conversation_history += f"\nUser: {user_input}\nBot:"
    response = chatbot(conversation_history, max_length=200, num_return_sequences=1, pad_token_id=50256)
    bot_reply = response[0]['generated_text'].split("Bot:")[-1].strip()
    conversation_history += f" {bot_reply}"
    return bot_reply

# Step 8: Gradio UI
interface = gr.Interface(fn=chat_with_bot,
                         inputs="text",
                         outputs="text",
                         title="College & Library Assistant Chatbot",
                         description="Ask about library, exams, assignments, reminders, or books.")
interface.launch()
```
